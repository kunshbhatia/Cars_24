```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cars_24.csv")
```

```python
df
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)	
```

```python
df['Year'] = df['Year'].fillna(2018.0) #Already checked using df['Year'].mode()[0]
df['Year'] = df['Year'].astype(int)
```

```python
df['Type'] = df['Type'].replace({'Lux_SUV':'Luxury SUV'})
df['Type'] = df['Type'].replace({"Lux_sedan" : "Luxury Sedan"})
```

```python
df.isnull().sum()
```

```python
df['Car Name'] = df['Car Name'].astype(str)
df['Location'] = df['Location'].astype(str)
```

```python
lst = []
for i in range(len(df['Car Name'])):
    temp = df['Car Name'][i].split()[1:]
    lst.append(" ".join(temp))

df['Name'] = np.array(lst)
```

```python
lst = []
lst1 = []
for i in range(len(df['Car Name'])):
    lst.append(df['Car Name'][i].split()[0])
    lst1.append(df['Location'][i].split("-")[0])

df['Brand'] = np.array(lst)
df['Location'] = np.array(lst1)
```

```python
df.loc[df['Location'] == '22',"Location"] = "MH"
df.loc[df['Location'] == 'nan',"Location"] = "MH"
```

```python
df.drop(['Car Name'],axis=1,inplace=True)
```

```python
df['Location'] = df['Location'].fillna("MH")
```

```python
df[['Year','Owner',"Fuel","Location","Drive",'Type']] = df[['Year','Owner',"Fuel","Location","Drive",'Type']].astype(str)
```

```python
df.drop(4901,inplace=True)
```

```python
df['Year'].unique()
```

```python
X = df[["Name",'Brand','Year','Distance','Owner','Fuel','Location','Drive','Type']]
Y = np.log1p(df['Price'])

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30)
```

```python
X_test
```

```python
#0 :- Auto 1 :- Manual
X_train['Drive'] = X_train['Drive'].apply( lambda x : 0 if x == "Automatic" else 1)
X_test['Drive'] = X_test['Drive'].apply( lambda x : 0 if x == "Automatic" else 1)
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler,OneHotEncoder
encode_year = OrdinalEncoder(categories=[["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021","2022",'2023']])
encode_distance = StandardScaler()
encode_fuel = OrdinalEncoder(categories=[['LPG','CNG','DIESEL','PETROL']])
encode_type = OrdinalEncoder(categories=[['HatchBack', 'Sedan', 'SUV', 'Luxury Sedan', 'Luxury SUV']])
encode_location = OrdinalEncoder(categories=[['HR', 'BR', 'WB', 'UP', 'MP', 'PB', 'RJ', 'CH', 'TN', 'TS', 'AP', 'KL', 'KA', 'GJ', 'MH', 'DL']])
encode_brand  = OrdinalEncoder(categories=[['Datsun', 'Renault',"Honda", 'Tata', 'Maruti', 'Hyundai', 'KIA', 'Ford', 'Nissan','Mahindra', 'Toyota', 'Volkswagen', 'Skoda', 'MG', 'Jeep', 'BMW']])
encode_name = OrdinalEncoder(categories=[a])

X_train[['Year']] = encode_year.fit_transform(X_train[['Year']])
X_test[['Year']] = encode_year.transform(X_test[['Year']])
X_train[['Distance']] = encode_distance.fit_transform(X_train[['Distance']])
X_test[['Distance']] = encode_distance.transform(X_test[['Distance']])
X_train[['Fuel']] = encode_fuel.fit_transform(X_train[['Fuel']])
X_test[['Fuel']] = encode_fuel.transform(X_test[['Fuel']])
X_train[['Type']] = encode_type.fit_transform(X_train[['Type']])
X_test[['Type']] = encode_type.transform(X_test[['Type']])
X_train[['Location']] = encode_location.fit_transform(X_train[['Location']])
X_test[['Location']] = encode_location.transform(X_test[['Location']])
X_train[['Brand']] = encode_brand.fit_transform(X_train[['Brand']])
X_test[['Brand']] = encode_brand.transform(X_test[['Brand']])
X_train['Name'] = encode_name.fit_transform(X_train[['Name']])
X_test['Name'] = encode_name.transform(X_test[['Name']])
```

```python
# Checked using all Regression Models , and found this as the best and added Hyperparameters . Want to see how then check my other ML Model for the code and all.
from sklearn.ensemble import RandomForestRegressor
regg = RandomForestRegressor(n_estimators= 200, max_depth= 30, criterion= 'friedman_mse')
regg.fit(X_train,Y_train)
```

```python
Y_predd = regg.predict(X_test)
Y_test_actual = np.expm1(Y_test)
Y_predd = np.expm1(Y_predd)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
score1 = r2_score(Y_test_actual,Y_predd)
mae1 = mean_absolute_error(Y_test_actual,Y_predd)
mse1 = mean_squared_error(Y_test_actual,Y_predd)
rmse = root_mean_squared_error(Y_test_actual,Y_predd)
print(f"R2 : {score1*100}")
print("MAE :",mae1)
print("MSE :",mse1)
print("RMSE :",rmse)

# Model is quite accurate for high end cars and for recent cars
```
